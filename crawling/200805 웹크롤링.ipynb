```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001'
```

```python
html = requests.get(url).text ;html
```

```python
soup = BeautifulSoup(html, 'html.parser') ;soup
```

```python
all = soup.find('ul',{'class':'type06_headline'})
```

```python
len(all)
```

```python
dl = all.find_all('dl') ; dl     ###list형 함수
```

```python
for item2 in dl:
    try:
        img=item2.find('dt',{'class':'photo'}).find('img')
        print('img=',img['src'])
        print('----------------------------')
    except:
        print('No image')
```

```python
dl = all.find_all('dl')
for item2 in dl:
    link = item2.find('dt',{'class':''}).find('a')
    print(link['href'])
    print(link.text.replace('\t','').replace('\n','')[1:len(link.text)+1])
                ##### replace= \t를 ''로 바꾸겠다, \n을 ''로 바꾸겠다.
                  ####  1부터 link.text값에 1더한 범위까지를 프린트하라.
```

```python
dl = all.find_all("dl")
for item2 in dl:
    try:
        content = item2.find("dd")
        print(content.text.replace("\t","").replace("\n","").split("…")[0])
    except:
        print("No Content")
```
