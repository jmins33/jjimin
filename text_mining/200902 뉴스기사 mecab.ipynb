### 뉴스 기사 크롤링
- https://www.yna.co.kr/view/AKR20200204081000017?section=it/it

```python
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get('https://www.yna.co.kr/view/AKR20200204081000017?ection=it/it')
raw = req.text

html = BeautifulSoup(raw, 'html.parser')
```

```python
div = html.find('div',class_='story-news article')
```

```python
p_tag = div.find_all('p')
```

```python
content = ''
for i in p_tag:
    content +=i.text
content
```

### 텍스트 마이닝
- 토큰화

```python
# Mecab
from konlpy.tag import Mecab
# 선언
mecab= Mecab()
# 토큰화 : morphs
mecab_tokens = mecab.morphs(content)
print(mecab_tokens)
```

### 품사 부착

```python
mecabTag=[]
for token in mecab_tokens:
    mecabTag+=mecab.pos(token)
print(mecabTag)
```

### 불용어 처리

```python
from collections import Counter
Counter(mecabTag).most_common()
```

```python
#불용어 처리
stopPos = ['NNB','NNBC','NR','NP','VV','VA','MM','MAG','MAJ','VX','IC','JKS','JKC','JKG','JKO','JKB','JKV',
           'JKQ','JC','JX','EP','EF','EC','ETN','ETM','XPN','XSN','XSV','XSA','XR',
           'SF','SE','SSO','SSC','SC','SY','SH','SL','SN','VV+EP','VV+ETM']

stopWord =['하','의','을','은','는','이','가','로','게','갖','적','내','인','와','유','도','과','면','서울',
           '연합뉴스','홍','지인','기자','무단','전재','재','배포','금지','놓','수','함','등',
           '저작권자','된다','.co','.kr','2020','02','04','11','33','송고']

word=[]

for tag in mecabTag:
    if tag[1] not in stopPos:
        if tag[0] not in stopWord:
            word.append(tag[0])
print(word)
```
