### 코로나와_온라인쇼핑_네이버_뉴스_키워드추출
참고 - https://liveyourit.tistory.com/57

```python
from konlpy.tag import Okt
from collections import Counter
import csv
```

```python
filename = "corona.txt"
f = open(filename, 'r', encoding='cp949')
news = f.read()
```

```python
# okt 객체 생성
okt = Okt()
noun = okt.nouns(news)
for i, v in enumerate(noun):
    if len(v)<2:
        noun.pop(i)
        
count = Counter(noun)
f.close()
```

```python
#명사 빈도 카운트
noun_list = count.most_common(100)
for v in noun_list:
    print(v)
```

```python
# txt 파일에 저장
with open('noun_list.txt','w', encoding='cp949') as f:
    for v in noun_list:
        f.write(" ".join(map(str,v))) # 튜플 int 값을 str 타입으로 전환 후 조인
        f.write("\n")
```
