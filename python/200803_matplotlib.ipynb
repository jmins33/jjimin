```python
x = np.linspace(0, 2, 100)  ; x
```

```python
########## 방법 1
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
######### 방법2
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
def my_plotter(ax, data1, data2, param_dict):
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2 = np.random.randn(2, 100)
fig, ax = plt.subplots(1, 1)
my_plotter(ax, data1, data2, {'marker': 'x'})
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)   ### 1행 2열짜리 그림을 그리겠다
my_plotter(ax1, data1, data2, {'marker':'x'})
my_plotter(ax2, data1, data2, {'marker':'o'})
```

```python
def my_plotter(ax, data1, data2, param_dict):
    out = ax[0].plot(data1, **param_dict)
    out = ax[1].plot(data2, **param_dict)
    return out

data1, data2 = np.random.randn(2, 100)
fig, ax = plt.subplots(2)
my_plotter(ax, data1, data2, {'marker': 'x'})
```

```python
def my_plotter(ax, data1, data2, param_dict):
    out = ax[0][0].plot(data1, **param_dict)
    out = ax[1][0].plot(data2, **param_dict)
    return out

data1, data2 = np.random.randn(2, 100)
fig, ax = plt.subplots(2,2)
my_plotter(ax, data1, data2, {'marker': 'x'})
```

```python
def my_plotter(ax, data1, data2, param_dict):
    out = ax[0][1].plot(data1, **param_dict)
    out = ax[1][1].plot(data2, **param_dict)
    return out

data1, data2 = np.random.randn(2, 100)
fig, ax = plt.subplots(2,3)
my_plotter(ax, data1, data2, {'marker': 'o'})
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data, alpha=0.9)   ###  x값, y값 표시, c는 컬러, s는 사이즈,
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9,3))   ####너비, 높이

plt.subplot(131)   ###1행3열을 그리는데 1번을 쓰겠다.
plt.bar(names, values)
plt.subplot(132)   ###1행3열을 그리는데 2번을 쓰겠다.
plt.scatter(names, values)
plt.subplot(133)   ###1행3열을 그리는데 3번을 쓰겠다.
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)
```

```python
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])   ### x와 y 범위
plt.grid(True)
plt.show()
```

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )

plt.ylim(-2, 2)
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
