```python
res = urllib.request.urlopen(url)      ###### 그냥 공백이 나올 경우, res 복붙해서 붙여넣고 하면 됨.
xml_str = res.read().decode('utf-8') ; xml_str

import xmltodict  ### xml문자열을..json 형태로 변환
import json

json_str = json.dumps(xmltodict.parse(xml_str), indent=4, ensure_ascii=False)
####                                  xml문자열을 딕셔너리 형태로 바꿔줌

json_object = json.loads(json_str)  ####문자열을 json 으로 변환
json_object
```
